```python
#import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read tge csv data
all_students = pd.read_csv("data/students_länder.csv")
```

```python
all_students.head()
```

```python
#rename the columns
all_students.rename(columns={'Unnamed: 0': 'semester', 'Unnamed: 1':'federal_state', 'Unnamed: 2':'study_code', 'Unnamed: 3': 'study_subject', 'Total': 'total_males', 'Total.1': 'total_females', 'Total.2': 'total_all'}, inplace=True)
```

```python
#replace - with null values
all_students.replace('-', np.nan, inplace=True)
```

```python
# drop null values
all_students.dropna(inplace=True)
```

```python
all_students.head()
```

```python
#create criteria for new data frame(IT)
mask1 = all_students['study_subject'].str.contains('computer', case=False)
```

```python
#create criteria for new data frame(IT)
mask2 = all_students['study_subject'].str.contains('informatics', case=False)
```

```python
#merge the 2 criteria(IT)
mask = mask1 | mask2
```

```python
#create new data frame based on criteria
it_students = all_students[mask]
```

```python
it_students.head()
```

```python
it_students['study_subject'].unique()
```

```python
it_students.head()
```

```python
#create criteria for new data frame(IT)
it = it_students['study_subject'].str.match('Computer Science')
```

```python
it_final = it_students[it]
```

```python
it_final
```

```python
#create csv file from data frame(IT students)
it_students.to_csv("data/it_students.csv", index=False)
```

```python
it_final.to_csv("data/it_narrowed.csv", index=False)
```
