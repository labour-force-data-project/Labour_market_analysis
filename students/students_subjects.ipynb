```python
#create criteria for new data frame
mask1 = students_subjects['study_field'].str.contains('informatik', case=False)
```

```python
mask2 = students_subjects['study_field'].str.contains('bauingenieurwesen', case=False)
```

```python
mask3 = students_subjects['study_field'].str.contains('architektur', case=False)
```

```python
mask4 = students_subjects['study_field'].str.contains('elektrotechnik', case=False)
```

```python
#merge all criteria into 1
mask = mask1 | mask2 | mask3 | mask4
```

```python
#create new data frame based on criteria
needed_students = students_subjects[mask]
```
