```python
#turn data type to integer when necessary
merged_data['federal_state_ID'] = pd.to_numeric(merged_data['federal_state_ID'], errors='coerce')
merged_data['professional_area_sub_ID'] = pd.to_numeric(merged_data['professional_area_sub_ID'], errors='coerce')
merged_data['unemployed'] = pd.to_numeric(merged_data['unemployed'], errors='coerce')
merged_data['job _seekers'] = pd.to_numeric(merged_data['job _seekers'], errors='coerce')
merged_data['reported_jobs_stock'] = pd.to_numeric(merged_data['reported_jobs_stock'], errors='coerce')
```

```python
#mean of unemployment
average_unemployed=merged_data['unemployed'].mean()
```

```python
average_unemployed
```

```python
#get values above the mean
high_unemployment = merged_data[merged_data["unemployed"]> average_unemployed]
```

```python
high_unemployment.describe()
```

```python
#group values by federal states
by_state = high_unemployment.groupby('federal_state')
result = by_state['unemployed'].sum()
result = result.sort_values(ascending=False)
print(result)
```

```python
#group values subcategory
by_cat = high_unemployment.groupby('professional_area_sub')
result1 = by_cat['unemployed'].sum()
result1 = result1.sort_values(ascending=False)
print(result1)
```

```python
#group job availabilities by federal states
by_state = high_unemployment.groupby('federal_state')
jobs = by_state['reported_jobs_stock'].sum()
jobs = jobs.sort_values(ascending=False)
print(jobs)
```

```python
#group jobs by subcategory
by_cat = high_unemployment.groupby('professional_area_sub')
jobs2 = by_cat['reported_jobs_stock'].sum()
jobs2 = jobs2.sort_values(ascending=False)
print(jobs2)
```
